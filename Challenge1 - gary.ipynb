```python
# Write a program to work out how many days you have lived for.
import datetime
today = datetime.date.today()
birthday = datetime.date(1998,11,5)
days = today - birthday
day = days.days

print("Days you have lived for: ",day)
```

```python
# Write a program to work out how many seconds you have lived for.
import datetime
today = datetime.date.today()
birthday = datetime.date(1998,11,5)
days = today - birthday
second = days.total_seconds()
print("Seconds you have lived for:",second)
```

```python
# Write a program that will accept a date of birth and determine whether one 
# is considered an adult (i.e. 21 years old).
import datetime
today = datetime.date.today()
birthday = datetime.date(1998,11,5)
days = (today - birthday).days
adult = 21 * 365
if days >= adult:
    print("adult")
else:
    print("not adult")

```

```python
# For the rock, paper, scissors game above, ensure that the user enters a valid 
# entry. Add a loop structure to play several times and keep a running score.

import random

while True:
    choice = input("Your choice: ")
    computer = random.randint(1,3)
    ans = {'rock':1,'paper':2,'scissors':3}
    if choice == 'end':
        break
    elif choice not in ans:
        print("Choose valid one pls")
    elif ans[choice] == computer:
        print("Draw")
    elif (ans[choice] == 1 and computer == 2):
        print("You Lose")
    elif (ans[choice] == 2 and computer == 3):
        print("You Lose")
    elif (ans[choice] == 3 and computer == 1):
        print("You Lose")
    else:
        print("You Win")
```

```python
# Write a program that will give the answer to logic gate questions e.g.
# Enter logic gate : OR
# Enter first input : 1
# Enter second input :0
# Result = 1
# It should work for the logic gates OR, XOR, NAND and NOR gates.

number = int(input("Enter number: "))

factors = []
for i in range(2,int(number**0.5+1)):
    if number % i == 0:
        if i == int(number/i): 
            factors += [i]
        else:
            factors += [i,int(number/i)]

        
print(sorted(factors))
```

```python
# Write a program that will display all the factors of an input number that
# are bigger than 1.
# (e.g. the factors of the number 12 are 6,4,3 and 2 because they divide
# into 12 exactly).
# Hint: To find out whether a number X is a factor of Y use:
# If Y mod X == 0 (there is nothing remaining when Y is divided by X)

num = input("Enter number:")
def factor(num):
    num = int(num)
    array = []
    for i in range(2,num):
        if num % i == 0:
            array.append(i)
    return array
print(factor(num))
```
